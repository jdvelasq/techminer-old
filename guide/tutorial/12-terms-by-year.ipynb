# Analysis: Terms by year

```python
import pandas as pd
from techminer import RecordsDataFrame

rdf = RecordsDataFrame(pd.read_json("step-07.json", orient="records", lines=True))
```

## Document type

```python
rdf.terms_by_year(column="Document Type")
```

```python
rdf.terms_by_year(column="Document Type", sep=None).tomatrix()
```

```python
rdf.terms_by_year(column="Document Type").heatmap(figsize=(5, 9))
```

## Source title

```python
rdf.terms_by_year(column="Source title", sep=None, top_n=20)
```

```python
rdf.terms_by_year(column="Source title", sep=None, top_n=20).tomatrix()
```

```python
rdf.terms_by_year(column="Source title", sep=None, top_n=10).heatmap(figsize=(8, 9))
```

```python
rdf.terms_by_year(column="Source title", sep=",", top_n=20).heatmap(figsize=(11, 8))
```

```python
rdf.terms_by_year(column="Source title", sep=",", top_n=30).heatmap(figsize=(11, 6))
```
